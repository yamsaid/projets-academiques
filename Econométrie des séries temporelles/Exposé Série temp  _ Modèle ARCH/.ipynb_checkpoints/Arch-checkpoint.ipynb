#  2. Nettoyage & traitement

```python
data['Prix'] = pd.to_numeric(data['Prix'], errors='coerce')
data['Temps'] = pd.to_datetime(data['Temps'], format='%Y-%m-%d')
```

# 3. Moyenne, écart-type, coefficient de variation

```python
mean_price = data['Prix'].mean()
sd_price = data['Prix'].std()
coeff_variation = sd_price / mean_price

print(f"La moyenne des prix des actions d'Apple est : {mean_price}")
print(f"L'écart-type des prix des actions d'Apple est : {sd_price}")
print(f"Le coefficient de variation est : {coeff_variation}")
```

 # 4. Visualisation

```python
import matplotlib.pyplot as plt

# Évolution du prix
plt.figure()
plt.plot(data['Temps'], data['Prix'])
plt.xlabel("Temps")
plt.ylabel("Prix des actions")
plt.title("Évolution des prix des actions d'Apple")
plt.show()

# Histogramme
plt.figure()
plt.hist(data['Prix'].dropna(), bins=30)
plt.title("Histogramme des prix des actions d'Apple")
plt.xlabel("Prix")
plt.ylabel("Fréquence")
plt.show()

# Densité
from scipy.stats import gaussian_kde
plt.figure()
density = gaussian_kde(data['Prix'].dropna())
xs = sorted(data['Prix'].dropna())
plt.plot(xs, density(xs))
plt.title("Densité des prix des actions d'Apple")
plt.xlabel("Prix")
plt.ylabel("Densité")
plt.show()
```

# 5. Test de stationnarité - ADF

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(data['Prix'].dropna())
print("ADF Statistic:", result[0])
print("p-value:", result[1])
```

#  6. Test de non-linéarité - BDS

```python
from statsmodels.tsa.stattools import bds

bds_result = bds(data['Prix'].dropna())
print(bds_result)
```

# 7. Carré de la série

```python
data1 = data.copy()
data1['Prix'] = data1['Prix'] ** 2

plt.figure()
plt.plot(data1['Temps'], data1['Prix'])
plt.title("Évolution des prix au carré des actions d'Apple")
plt.xlabel("Temps")
plt.ylabel("Prix²")
plt.show()
```

# 8. Rendement

```python
# from pandas_datareader._utils import RemoteDataError
import numpy as np

data['Rendement'] = data['Prix'].pct_change()

plt.figure()
plt.plot(data['Temps'], data['Rendement'])
plt.title("Rendement des actions")
plt.xlabel("Temps")
plt.ylabel("Rendement")
plt.show()
```

# 9. Différenciation + Test ADF

```python
diff_series = data1['Prix'].diff().dropna()
adf_diff = adfuller(diff_series)
print("ADF sur différenciée :", adf_diff[0], "p-value:", adf_diff[1])
```

# 10. ACF / PACF

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(diff_series)
plt.show()

plot_pacf(diff_series)
plt.show()
```

# 11. Transformation logarithmique + test KPSS

```python
from statsmodels.tsa.stattools import kpss

data1 = data.copy()
data1['Prix'] = np.log(data1['Prix'])

kpss_stat, p_value, _, _ = kpss(data1['Prix'].dropna(), regression='c')
print("KPSS Statistic:", kpss_stat)
print("p-value:", p_value)

plt.figure()
plt.plot(data1['Temps'], data1['Prix'])
plt.title("Série transformée en log")
plt.show()
```

#  12. Correlogramme

```python
plot_acf(data1['Prix'].dropna())
plt.show()

plot_pacf(data1['Prix'].dropna())
plt.show()
```

# 13. Modèle ARCH(1)

```python
from arch import arch_model

# ARCH(1) - sans moyenne (mean='Zero')
returns = data1['Prix'].dropna().pct_change().dropna() * 100
model = arch_model(returns, vol='ARCH', p=1, mean='Zero')
res = model.fit()
print(res.summary())
```

#  14. Prévisions

```python
forecast = res.forecast(horizon=10)
print(forecast.variance[-1:])

plt.figure()
plt.plot(forecast.variance.values[-1])
plt.title("Prévision de la variance")
plt.show()
```
